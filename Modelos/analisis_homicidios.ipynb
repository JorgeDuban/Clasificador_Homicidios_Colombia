```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/content/drive/MyDrive/bigdata3/proyecto homicidios/Homicidios.xlsx')
```

```python
# -- Conociendo las estructura de los datos
print("\n \t En el dataset hay {0[0]} filas y {0[1]} columnas. \n".format(data.shape))
data.info()
data.head()
print(data[0:3])
print(pd.isnull(data).sum())
print(data.describe())
```

```python
# -- Visualización de los datos
diagnosis_all = list(data.shape)[0]
print(diagnosis_all)
diagnosis_categories = list(data['Edad'].value_counts())
print(diagnosis_categories)
print("\n \t En los datos hay {} dignosticos, {} malignos {} benignos.".format(diagnosis_all, 
                                                                                 diagnosis_categories[0], 
                                                                                 diagnosis_categories[1]))
```

```python
# Se extraen las culumnas que tienen las medias
features_mean= list(data.columns[2:12])
```

```python
#Se muestra la correlación ntre variables
plt.figure(figsize=(10,10))
sns.heatmap(data[features_mean].corr(), annot=True, square=True, cmap='coolwarm')
plt.show()
```

```python
# Valores de correlación
corr_pairs = data[features_mean].corr().unstack()
print(corr_pairs)
```

```python
# Pintar la ubicación de la variable segun el diagnóstico
color_dic = {'M':'red', 'B':'blue'}
colors = data['Edad'].map(lambda x: color_dic.get(x))
sm = pd.scatter_matrix(data[features_mean], c=colors, alpha=0.4, figsize=((15,15)))
plt.show()
```

```python
# Pintar valores del dignóstico según cada variable media
bins = 12
plt.figure(figsize=(15,15))
for i, feature in enumerate(features_mean):
    rows = int(len(features_mean)/2)    
    plt.subplot(rows, 2, i+1)    
    sns.distplot(data[data['Edad']=='M'][feature], bins=bins, color='red', label='M')
    sns.distplot(data[data['Edad']=='B'][feature], bins=bins, color='blue', label='B')    
    plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Una forma de ver (comun en análisis de bolsa de valores)
plt.figure(figsize=(15,15))
for i, feature in enumerate(features_mean):
    rows = int(len(features_mean)/2)
    
    plt.subplot(rows, 2, i+1)
    
    sns.boxplot(x='Edad', y=feature, data=data, palette="Set1")

plt.tight_layout()
plt.show()
```
